# Estudio de las polarizaciones y las transformaciones generadas por la rotación de placas de retardo de $ \ \frac{\lambda}{4} $

## Polarización

***Definición de polarización***

Sea $ \ \mathbb{P} = \{ X : \mathbb{R} \to \mathbb{R}^2 \ / \ \exists \ \vec{u},\ \vec{v} \in \mathbb{R}^2 \ \land \ X(t) = \vec{u} \cos{t} + \vec{v} \sin{t} \} $ 

Decimos que $ \ X \ $ es una polarización $ \iff X \in \mathbb{P} $ 

Es fácil ver que si $ \ X \ $ es una polarización, entonces $ \ X(t) = X(t + 2\pi) \quad \forall t \in \mathbb{R} $

<br>

***Unicidad de $ \ \vec{u}\ $ y $ \ \vec{v} $***

Veamos que si $\ X \in \mathbb{P}, \ $ entonces $ \ \exists! \ \vec{u},\ \vec{v} \in \mathbb{R}^2 \ / \ X(t) =  \vec{u} \cos{t} + \vec{v} \sin{t} \\ $ 
Supongamos que existen $ \ \vec{u}_1, \ \vec{v}_1 \ $ y $ \ \vec{u}_2, \ \vec{v}_2 \ $ tales que:
$$  X(t) = \vec{u}_1 \cos{t} + \vec{v}_1 \sin{t} = \vec{u}_2 \cos{t} + \vec{v}_2 \sin{t} \quad \forall t \in \mathbb{R} $$
Entonces, usando $ \ t = 0, \ $ se tiene $ \vec{u}_1 = \vec{u}_2 \ $ y con $ t = \frac{\pi}{2} \ $ resulta $ \ \vec{v}_1 = \vec{v}_2 \quad \square $ 

<br>

***Polarizaciones equivalentes***

Sea $ \ X_1 \in \mathbb{P} \ $ tal que $ \ X_1(t) = \vec{u}_1 \cos{t} + \vec{v}_1 \sin{t} \ $ y consideremos la siguiente función $ \ X_2 : \mathbb{R} \to \mathbb{R}^2 \ $ tal que $ \ X_2(t) = X_1(t + t_0) \ $ con un $ \ t_0 \in \mathbb{R} \ $ cualquiera

Veamos que $ \ X_2 \in \mathbb{P}: \ $

$$ X_2(t) = \vec{u}_1 \cos{(t + t_0)} + \vec{v}_1 \sin{(t + t_0)} = \vec{u}_1 ( \cos{t} \cos{t_0} - \sin{t} \sin{t_0} ) + \vec{v}_1 ( \cos{t} \sin{t_0} + \sin{t} \cos{t_0} )$$

Reordenando, se obtiene :
$$ X_2(t) = ( \vec{u}_1 \cos{t_0} + \vec{v}_1 \sin{t_0} ) \cos{t} + ( \vec{v}_1 \cos{t_0} - \vec{u}_1 \sin{t_0} ) \sin{t} $$

tomando $ \ \vec{u}_2 = \vec{u}_1 \cos{t_0} + \vec{v}_1 \sin{t_0} \ $ y $ \ \vec{v}_2 = \vec{v}_1 \cos{t_0} - \vec{u}_1 \sin{t_0} \ $ resulta :

$$ X_2(t) = \vec{u}_2 \cos{t} + \vec{v}_2 \sin{t} \quad \land \quad \vec{u}_2,\ \vec{v}_2 \in \mathbb{R}^2 $$

por lo tanto $ \ X_2 \in \mathbb{P} \quad \square $

<br>

Definamos una relación en $ \ \mathbb{P} \ $ de la siguiente manera:

Sean $ \ X_1,\ X_2 \in \mathbb{P} \ $ dos polarizaciones, decimos que $ \ X_1 \ $ es equivalente a $ \ X_2 \ $ y lo notamos $ \ X_1 \sim X_2 \iff \exists \ t_0 \in \mathbb{R} \ / \ X_2(t) = X_1(t + t_0) \quad \forall t \in \mathbb{R} $

Veamos que esta relación es de equivalencia en $ \ \mathbb{P}, \ $ es decir que verifica las propiedades de Reflexividad, Simetria y Transitividad

*Reflexividad*

Observando que:
$$ \ X_1(t) = X_1(t + 0) \quad \forall t \in \mathbb{R} $$ 

y tomando $ \ t_0 = 0, \ $ resulta, por definición, que $ \ X_1 \sim X_1 \ $  

*Simetria*

Probemos que $ \ X_1 \sim X_2 \implies X_2 \sim X_1 \ $

$$ \ X_1 \sim X_2 \implies \exists \ t_0 \in \mathbb{R} \ / \ X_2(t) = X_1(t + t_0) \quad \forall t \in \mathbb{R} $$

Entonces, remplazando $ \ t \ $ por $ \ t - t_0 \ $ :
$$ \ X_2(t - t_0) = X_1((t - t_0) + t_0) = X_1(t) \ $$

Luego, tomando $ \ t_1 = -t_0, \ $ podemos decir que $ \ \exists \ t_1 \in \mathbb{R} \ $ tal que $ \ X_1(t) = X_2(t + t_1) \quad \forall t \in \mathbb{R} $

y por definición, se tiene que $ \ X_2 \sim X_1 \ $


*Transitividad*

Probemos que $ \ X_1 \sim X_2 \ \land \ X_2 \sim X_3 \implies X_1 \sim X_3 $

$ \ X_1 \sim X_2 \ \land \ X_2 \sim X_3 \implies \exists \ t_1, t_2 \in \mathbb{R} \ / \ X_2(t) = X_1(t + t_1) \ \land \ X_3(t) = X_2(t + t_2)$

Remplazando $ \ t \ $ por $ \ t + t_2 \ $ :

$$ X_2(t + t_2) = X_1((t + t_2) + t_1) = X_1(t + (t_1 + t_2)) $$

y por lo tanto:

$$ X_3(t) = X_1(t + (t_1 + t_2)) $$

tomando $ \ t_0 = t_1 + t_2 \ $ en la definición de la relación, se tiene que $ \ X_1 \sim X_3 \quad \square$  

<br>

***Otra caracterización para polarizaciones equivalentes***

Sean $ X_1,\ X_2 \in \mathbb{P} \ $ dos polarizaciones tales que $ \ X_1(t) = \vec{u}_1 \cos{t} + \vec{v}_1 \sin{t} \ $ y $ \ X_2(t) = \vec{u}_2 \cos{t} + \vec{v}_2 \sin{t} \ $ 

entonces, utilizando la definición de equivalencia entre polarizaciones, la unicidad de $ \ \vec{u} \ $ y $ \ \vec{v} \ $ para una polarización dada, y que, como se mostró más arriba :

$$ X_1(t + t_0) = ( \vec{u}_1 \cos{t_0} + \vec{v}_1 \sin{t_0} ) \cos{t} + ( \vec{v}_1 \cos{t_0} - \vec{u}_1 \sin{t_0} ) \sin{t} $$

se tiene que :
$$ X_1 \sim X_2 \iff \exists \ t_0 \in \mathbb{R} \ / \ \vec{u}_2 = \vec{u}_1 \cos{t_0} + \vec{v}_1 \sin{t_0} \ \land \ \vec{v}_2 = \vec{v}_1 \cos{t_0} - \vec{u}_1 \sin{t_0} $$

<br>

***Igualdad entre las imagenes de polarizaciones equivalentes***

Ahora veamos que las imagenes de dos polarizaciones equivalentes son iguales, esto es, dadas dos polarizaciones $ \ X_1, X_2 \in \mathbb{P} : $

$$ X_1 \sim X_2 \implies Im(X_1) = Im(X_2) $$

Veamos que $ \ Im(X_1) \subseteq Im(X_2) \ $ :

Sea $ \ \vec{p} \in Im(X_1) \implies \exists \ t_1 \in \mathbb{R} \ / \ \vec{p} = X_1(t_1) $

Usando la simetria de la relación tenemos que $ \ X_1 \sim X_2 \implies X_2 \sim X_1 \ $ y por lo tanto, existe $ \ t_0 \in \mathbb{R} \ $ tal que:

$$ X_1(t) = X_2(t + t_0) \quad \forall t \in \mathbb{R} $$ 

Evaluando en $ \ t = t_1 \ $ 

$$ \vec{p} = X_1(t_1) = X_2(t_1 + t_0) $$ 

Como $ \ t_0 + t_1 \in \mathbb{R}, \ $ se tiene que $ \ \vec{p} \in Im(X_2) $

La otra inclusión se prueba de manera análoga.

**Polarizaciones principales**

Sea $ \ X \in \mathbb{P} \ $ una polarización tal que $ \ X(t) = \vec{u} \cos{t} + \vec{v} \sin{t} $

Diremos que $ \ X \ $ es una polarización principal $ \iff \langle \vec{u},\ \vec{v} \rangle = 0 \ $ y se cumple alguna de las dos condiciones siguientes: 

1. $ \ \lVert \vec{u}_p \rVert > \lVert \vec{v}_p \rVert \ $ y que el ángulo, $\ \alpha, \ $ entre $ \ \vec{u}_p \ $ y el eje $\ x \ $ sea tal que $ \ 0 \leq \alpha < \pi \ $ 
2. $ \ \vec{u}_p \parallel \begin{pmatrix} 1 \\ 0 \end{pmatrix} \ $


Dada una polarización cualquiera $\ X, \ $ veamos como hallar una polarizacion equivalente y que sea principal, para ello 

debemos encontrar un $ \ t_0 \ $ tal que:

$$ \langle \vec{u} \cos{t_0} + \vec{v} \sin{t_0}, \ \vec{v} \cos{t_0} - \vec{u} \sin{t_0} \rangle = 0 $$

Utilizando las propiedades del producto interno, se tiene :

$$ \langle \vec{u}, \ \vec{v} \rangle \ (\ \cos^2{t_0} - \sin^2{t_0} \ ) +  \cos{t_0} \sin{t_0} \ ( \ \lVert \vec{v} \rVert^2 - \lVert \vec{u} \rVert^2 \ ) = 0 $$

$$ \langle \vec{u}, \ \vec{v} \rangle \ \cos{2t_0} + \frac{1}{2} \sin{2t_0} \ ( \ \lVert \vec{v} \rVert^2 - \lVert \vec{u} \rVert^2 \ ) = 0 $$

* Si $ \ \lVert \vec{u} \rVert \neq \lVert \vec{v} \rVert \ $ :


$$ \tan{2t_0} = - \frac{2 \ \langle \vec{u}, \ \vec{v} \rangle}{\lVert \vec{v} \rVert^2 - \lVert \vec{u} \rVert^2} $$
$$ t_p = \frac{1}{2} \arctan{(- \frac{2 \ \langle \vec{u}, \ \vec{v} \rangle}{\lVert \vec{v} \rVert^2 - \lVert \vec{u} \rVert^2})} $$

$$ t_0 = t_p + k \frac{\pi}{2}, \quad k \in \{0,1,2,3\} $$

Tomando $ \ X_p(t) = \vec{u}_p \cos{t} + \vec{v}_p \sin{t} \ \ $ con : 

$\vec{u}_p = \vec{u} \cos{t_0} + \vec{v} \sin{t_0} \ $ y $ \ \vec{v}_p = \vec{v} \cos{t_0} - \vec{u} \sin{t_0} \ $

se tiene que $ \ X \sim X_p \ \land \langle \vec{u}_p,\ \vec{v}_p \rangle = 0 $ 

Los 4 pares de valores posibles son :

$$ \vec{u}_p = \vec{u} \ ( \ \cos{(k \frac{\pi}{2})} \cos{t_p} - \sin{(k \frac{\pi}{2})} \sin{t_p} \ ) + \vec{v} \ ( \ \cos{(k \frac{\pi}{2})} \sin{t_p} + \sin{(k \frac{\pi}{2})} \cos{t_p} \ ) $$ 

$$ \vec{v}_p = \vec{v} \ ( \ \cos{(k \frac{\pi}{2})} \cos{t_0} - \sin{(k \frac{\pi}{2})} \sin{t_p} \ ) - \vec{u} \ ( \ \cos{(k \frac{\pi}{2})} \sin{t_p} + \sin{(k \frac{\pi}{2})} \cos{t_p} \ ) $$ 

$$ \vec{u}_p = \begin{cases} \vec{u} \cos{t_p} + \vec{v} \sin{t_p} \\ \vec{v} \cos{t_p} - \vec{u} \sin{t_p} \\ -\vec{u} \cos{t_p} - \vec{v} \sin{t_p} \\ - \vec{v} \cos{t_p} + \vec{u} \sin{t_p} \end{cases} \quad \quad \vec{v}_p = \begin{cases} \vec{v} \cos{t_p} - \vec{u} \sin{t_p} & k = 0 \\ -\vec{u} \cos{t_p} - \vec{v} \sin{t_p} & k = 1 \\ -\vec{v} \cos{t_p} + \vec{u} \sin{t_p} & k = 2 \\ \vec{u} \cos{t_p} + \vec{v} \sin{t_p} & k = 3 \end{cases} $$

* Si $ \ \lVert \vec{u} \rVert = \lVert \vec{v} \rVert \ $ :

$$ \langle \vec{u}, \ \vec{v} \rangle \ \cos{2t_0} = 0 $$

Si $ \ \langle \vec{u}, \ \vec{v} \rangle \neq 0 \ $ entonces :
$$ t_0 = \frac{\pi}{4} + k \frac{\pi}{2} \quad k \in \{0, 1, 2, 3\} $$

Si $ \ \langle \vec{u}, \ \vec{v} \rangle = 0, \ $ que es el caso circular, entonces, $ \ t_0 \in \mathbb{R} $

Se puede probar que, salvo en el caso circular, existe una única solución tal que $\ \lVert \vec{u}_p \rVert > \lVert \vec{v}_p \rVert \ $ y que el ángulo, $\ \alpha, \ $ entre $ \ \vec{u}_p \ $ y el eje $\ x \ $ sea tal que $ \ 0 \leq \alpha < \pi $

En el caso circular se puede tomar $ \ \vec{u}_p \parallel \begin{pmatrix} 1 \\ 0 \end{pmatrix} \ $

Teniendo en cuenta estas condiciones, podemos decir que dada una polarización cualquiera, existe una única polarización equivalente y principal.

Por lo tanto se puede identificar a cada clase de equivalencia del conjunto cociente $ \ \mathbb{P} \ / \sim, \ $ con la polarizacion principal de cada clase.

Si $\ \vec{u} \ $ y $\ \vec{v} \ $ coinciden con los ejes principales de la elipse, como muestra la figura, estos se pueden caracterizar en terminos de $ r, \alpha $ y $ \beta $ de la siguiente manera:

**$\mathbb{P} \ $ como espacio vectorial**

Es fácil ver que $ \ \mathbb{P} \ $ junto a la suma usual de funciones y el producto usual de funciones por escalares, tiene estructura de $ \ \mathbb{R} $-espacio vectorial.

Solo mostraremos que $ \ \mathbb{P} \ $ es cerrado para la suma y el producto por escalares.

Sean $ \ X_1, X_2 \in \mathbb{P} \ $ dos polarizaciones tales que $\ X_1(t) = \vec{u}_1 \cos{t} + \vec{v}_1 \sin{t} \ $ y $ \ X_2(t) = \vec{u}_2 \cos{t} + \vec{v}_2 \sin{t} \ $ 

Entonces por la suma usual de funciones, se tiene que:
$$ (X_1 + X_2)(t) = X_1(t) + X_2(t) \quad \forall \ t \in \mathbb{R} $$ 
$$ (X_1 + X_2)(t) = (\vec{u}_1 + \vec{u}_2) \cos{t} + (\vec{v}_1 + \vec{v}_2) \sin{t} $$

como $ \ \vec{u}_1 + \vec{u}_2 \ $ y $ \ \vec{v}_1 + \vec{v}_2 \in \mathbb{R}^2 \ $ entonces $ \ X_1 + X_2 \in \mathbb{P} $ 

Veamos el producto por escalares, sea $ \ k \in \mathbb{R} \ $ y $ \ X \in \mathbb{P}, \ $ entonces :

$$ (k \cdot X)(t) = k \cdot X(t) \quad \forall \ t \in \mathbb{R} $$ 
$$ (k \cdot X)(t) = k \cdot \vec{u} \cos{t} + k \cdot \vec{v} \sin{t} $$ 

como $ \ k \cdot \vec{u} \ $ y $ \ k \cdot \vec{v} \in \mathbb{R}^2 \ $ entonces $ \ k \cdot X \in \mathbb{P} \quad \square $ 

<br>

Probemos que $ \ \mathbb{P} \ $ es isomorfo a $ \ \mathbb{R}^4 \\ $

Consideremos $\ X \in \mathbb{P} \ / \ X(t) = \vec{u} \cos{t} + \vec{v} \sin{t} \ $ con $ \ \vec{u} = \begin{pmatrix} u_x \\ u_y \end{pmatrix} \ $ y $ \ \vec{v} = \begin{pmatrix} v_x \\ v_y \end{pmatrix} \ $

Sea $ \ f : \mathbb{P} \to \mathbb{R}^4 \ $ tal que $ \ f(X) = \begin{pmatrix} u_x \\ u_y \\ v_x \\ v_y \end{pmatrix} \ $ que está bien definida por la unicidad de $ \ \vec{u} \ $ y $ \ \vec{v} \ $

$ \ f \ $ es una transformación lineal, ya que :

$$ f(X_1 + X_2) = \begin{pmatrix} u_{1x} + u_{2x} \\ u_{1y} + u_{2y} \\ v_{1x} + v_{2x} \\ v_{1y} + v_{2y} \end{pmatrix} = \begin{pmatrix} u_{1x} \\ u_{1y} \\ v_{1x} \\ v_{1y} \end{pmatrix} + \begin{pmatrix} u_{2x} \\ u_{2y} \\ v_{2x} \\ v_{2y} \end{pmatrix} = f(X_1) + f(X_2) $$

$$ f(k \cdot X) = \begin{pmatrix} k \ u_x \\ k \ u_y \\ k \ v_x \\ k \ v_y \end{pmatrix} = k \cdot \begin{pmatrix} u_x \\ u_y \\ v_x \\ v_y \end{pmatrix} = k \cdot f(X) $$

Veamos que es biyectiva y por lo tanto un isomorfismo:

Primero mostremos que $ \ f \ $ es inyectiva:

$$ f(X_1) = f(X_2) \implies \vec{u}_1 = \vec{u}_2 \ \land \ \vec{v}_1 = \vec{v}_2 \implies X_1 = X_2 $$

Ahora veamos la suryectividad :

Sea $ \ \vec{y} = \begin{pmatrix} y_1 \\ y_2 \\ y_3 \\ y_4 \end{pmatrix} \in \mathbb{R}^4, \ $ entonces tomando $ \ X(t) = \begin{pmatrix} y_1 \\ y_2 \end{pmatrix} \cos{t} + \begin{pmatrix} y_3 \\ y_4 \end{pmatrix} \sin{t}, \ $ se tiene que $ X \in \mathbb{P} \ $ y $ \ f(X) = \vec{y} \quad \square $

<br>

Sea $ B = \{ \begin{pmatrix} 1 \\ 0 \end{pmatrix} \cos{t}, \ \begin{pmatrix} 0 \\ 1 \end{pmatrix} \cos{t}, \ \begin{pmatrix} 1 \\ 0 \end{pmatrix} \sin{t}, \begin{pmatrix} 0 \\ 1 \end{pmatrix} \sin{t} \}, \ $ probemos que es una base de $ \ \mathbb{P}, \ $ a la que llamaremos base canónica, 

para ello solo basta mostrar que sus elementos son l. i., ya que $ \ dim(\mathbb{P}) = 4 \ $ por ser isomorfo a $ \ \mathbb{R}^4 \ $

Veamos que son linealmente independientes.

Sean $ \ a,\ b,\ c,\ d \in \mathbb{R} \ $ tales que:
$$ a \ \begin{pmatrix} 1 \\ 0 \end{pmatrix} \cos{t} + b \ \begin{pmatrix} 0 \\ 1 \end{pmatrix} \cos{t} + c \  \begin{pmatrix} 1 \\ 0 \end{pmatrix} \sin{t} + d \ \begin{pmatrix} 0 \\ 1 \end{pmatrix} \sin{t} = \begin{pmatrix} 0 \\ 0 \end{pmatrix} \quad \forall t \in \mathbb{R} $$

entonces para $ \ t = 0, \ $ se tiene que $ \ \begin{pmatrix} a \\ b \end{pmatrix} = \begin{pmatrix} 0 \\ 0 \end{pmatrix}\ $ y para $ \ t = \frac{\pi}{2}, \ $ se tiene que $ \ \begin{pmatrix} c \\ d \end{pmatrix} = \begin{pmatrix} 0 \\ 0 \end{pmatrix}\ $ 

Por lo que resulta que $ \ a = b = c = d = 0 \quad \square $

<br>

La función $ \ f \ $ definida más arriba equivale a tomar coordenadas en la base $ \ B \ $ de una polarización dada $ \ X \ $.  




## Transformación

**Descripción de la rotación de la placa de retardo**

Ejes de la placa de retardo en función de la rotación $\theta \in [0, \pi]$

    <img src="retardo.png">
</div>

$$ \begin{aligned}
    Eje\ lento &:\vec{s}_\theta = \begin{pmatrix} \cos{\theta} \\ \sin{\theta} \end{pmatrix} \\ \\
    Eje\ rapido &:\vec{f}_\theta = \begin{pmatrix} -\sin{\theta} \\ \cos{\theta} \end{pmatrix} 
\end{aligned} $$
$$ \vec{s}_\theta \perp \vec{f}_\theta $$

**Definicion de la Transformacion $ \ T_\theta \ $**

$$ T_\theta : \mathbb{P} \to \mathbb{P} $$

<div align="center">
    <img src="transformacion.png">
</div>


Sea  $ \ X \in \mathbb{P} \ $ tal que $ \ X(t) = \vec{u} \cos{t} + \vec{v} \sin{t} \ $ 

Ahora consideremos la función $ \ Y : \mathbb{R} \to \mathbb{R}^2 \ $ tal que 
$$ Y(t) = \langle X(t),\ \vec{f}_{\theta} \rangle \ \vec{f}_{\theta} + \langle X(t - \frac{\pi}{2}),\ \vec{s}_{\theta} \rangle \ \vec{s}_{\theta} \ $$
Veamos que $ \ Y \in \mathbb{P} \ $ 

$$ Y(t) = \langle \vec{u} \cos{t} + \vec{v} \sin{t},\ \vec{f}_\theta \rangle \ \vec{f}_\theta + \langle -\vec{v} \cos{t} + \vec{u} \sin{t},\ \vec{s}_\theta \rangle \ \vec{s}_\theta$$

Escribiendo $ \ \vec{p} \ $ y $ \ \vec{q} \ $ de la siguiente manera:

$$ \vec{p} = \langle \vec{u},\ \vec{f}_\theta \rangle \ \vec{f}_\theta - \langle \vec{v},\ \vec{s}_\theta \rangle \ \vec{s}_\theta $$

$$ \vec{q} = \langle \vec{v},\ \vec{f}_\theta \rangle \ \vec{f}_\theta + \langle \vec{u},\ \vec{s}_\theta \rangle \ \vec{s}_\theta $$

y utilizando las propiedades del producto interno, se tiene que :

$$ Y(t) = \vec{p} \cos{t} + \vec{q} \sin{t} \quad \land \quad \vec{p},\ \vec{q} \in \mathbb{R}^2 $$

Luego, $ \ Y \in \mathbb{P} \quad \square $ 

<br>

Ahora definamos $ \ T_\theta : \mathbb{P} \to \mathbb{P} \ $ como : $$ \ T_\theta(X) = Y $$

Veamos que es una transformación lineal y luego daremos su matriz asociada en la base canónica de $ \ \mathbb{P} \ $

Probemos que, $\  T_\theta(X_1 + X_2) = T_\theta(X_1) + T_\theta(X_2) $ 

$$ T_\theta(X_1 + X_2)(t) = \langle (X_1 + X_2)(t),\ \vec{f}_\theta \rangle \ \vec{f}_\theta + \langle (X_1 + X_2)(t - \frac{\pi}{2}),\ \vec{s}_\theta \rangle \ \vec{s}_\theta = \langle X_1(t) + X_2(t),\ \vec{f}_\theta \rangle \ \vec{f}_\theta + \langle X_1(t - \frac{\pi}{2})) + X_2(t - \frac{\pi}{2}),\ \vec{s}_\theta \rangle \ \vec{s}_\theta $$

Utilizando las propiedades del producto interno y reordenando :

$$ T_\theta(X_1 + X_2)(t) = \langle X_1(t),\ \vec{f}_\theta \rangle \ \vec{f}_\theta + \langle X_1(t - \frac{\pi}{2}),\ \vec{s}_\theta \rangle \ \vec{s}_\theta + \langle X_2(t),\ \vec{f}_\theta \rangle \ \vec{f}_\theta + \langle X_2(t - \frac{\pi}{2}),\ \vec{s}_\theta \rangle \ \vec{s}_\theta = T_\theta(X_1)(t) + T_\theta(X_2)(t) $$

ahora veamos que, $ \ T_\theta(k \cdot X) = k \cdot T_\theta(X) $ 

$$ T_\theta(k \cdot X)(t) = \langle (k \cdot X)(t),\ \vec{f}_\theta \rangle \ \vec{f}_\theta + \langle (k \cdot X)(t - \frac{\pi}{2}),\ \vec{s}_\theta \rangle \ \vec{s}_\theta = \langle k \ X(t),\ \vec{f}_\theta \rangle \ \vec{f}_\theta + \langle k \ X(t - \frac{\pi}{2}),\ \vec{s}_\theta \rangle \ \vec{s}_\theta$$

$$ T_\theta(k \cdot X)(t) = k \cdot (\ \langle X(t),\ \vec{f}_\theta \rangle \ \vec{f}_\theta + \langle X(t - \frac{\pi}{2}),\ \vec{s}_\theta \rangle \ \vec{s}_\theta \ ) = k \cdot T_\theta(X) \quad \square $$

<br>

Sea $ \ B = \{ X_1,\ X_2,\ X_3,\ X_4 \} \ $ la base canonica de $\ \mathbb{P} \ $, entonces:

$$ X_1(t) = \begin{pmatrix} 1 \\ 0 \end{pmatrix} \cos{t} \quad X_2(t) = \begin{pmatrix} 0 \\ 1 \end{pmatrix} \cos{t} \quad X_3(t) = \begin{pmatrix} 1 \\ 0 \end{pmatrix} \sin{t} \quad X_4(t) = \begin{pmatrix} 0 \\ 1 \end{pmatrix} \sin{t} $$

Entonces

$$ T_\theta(X_1)(t) = \begin{pmatrix} \sin^2{\theta} \\ - \sin{\theta} \cos{\theta} \end{pmatrix} \cos{t} + \begin{pmatrix} \cos^2{\theta} \\ \sin{\theta} \cos{\theta} \end{pmatrix} \sin{t} $$

$$ T_\theta(X_2)(t) = \begin{pmatrix} - \sin{\theta} \cos{\theta} \\ \cos^2{\theta} \end{pmatrix} \cos{t} + \begin{pmatrix} \sin{\theta} \cos{\theta} \\ \sin^2{\theta} \end{pmatrix} \sin{t} $$

$$ T_\theta(X_3)(t) = \begin{pmatrix} - \cos^2{\theta} \\ - \sin{\theta} \cos{\theta} \end{pmatrix} \cos{t} + \begin{pmatrix} \sin^2{\theta} \\ - \sin{\theta} \cos{\theta} \end{pmatrix} \sin{t} $$

$$ T_\theta(X_4)(t) = \begin{pmatrix} - \sin{\theta} \cos{\theta} \\ - \sin^2{\theta} \end{pmatrix} \cos{t} + \begin{pmatrix} - \sin{\theta} \cos{\theta} \\ \cos^2{\theta} \end{pmatrix} \sin{t} $$

Por lo tanto la matriz asociada a la transformación:

$$ \lvert T_\theta \rvert _B = \begin{pmatrix} \sin^2{\theta} & - \sin{\theta} \cos{\theta} & - \cos^2{\theta} & - \sin{\theta} \cos{\theta} \\ - \sin{\theta} \cos{\theta} & \cos^2{\theta} & - \sin{\theta} \cos{\theta} & - \sin^2{\theta} \\ \cos^2{\theta} & \sin{\theta} \cos{\theta} & \sin^2{\theta} & - \sin{\theta} \cos{\theta} \\ \sin{\theta} \cos{\theta} & \sin^2{\theta} & - \sin{\theta} \cos{\theta} & \cos^2{\theta} \end{pmatrix} $$

<br>

Probemos lo siguiente:

$$ X_1 \sim X_2 \implies T_\theta(X_1) \sim T_\theta(X_2) $$

Como:
$$ X_1 \sim X_2 \implies \exists \ t_0 \ / \ X_2(t) = X_1(t + t_0) $$

Entonces:

$$ T_\theta(X_2)(t) = \langle X_2(t),\ \vec{f}_\theta \rangle \ \vec{f}_\theta + \langle X_2(t-\frac{\pi}{2}),\ \vec{s}_\theta \rangle \ \vec{s}_\theta = \langle X_1(t + t_0),\ \vec{f}_\theta \rangle \ \vec{f}_\theta + \langle X_1(t -\frac{\pi}{2} + t_0),\ \vec{s}_\theta \rangle \ \vec{s}_\theta = T_\theta(X_1)(t + t_0) \quad \square$$


```python
import numpy as np
from numpy.linalg import norm
import matplotlib.pyplot as plt
from ipywidgets import *
```

```python
dt = 0.01

def X(r, alpha, beta):
    """Descripcion de una polarizacion principal en funcion de r, a, b"""    
    
    ux = r * np.cos(alpha)
    uy = r * np.sin(alpha) 

    vx = r * np.tan(beta) * (- np.sin(alpha))
    vy = r * np.tan(beta) * np.cos(alpha)

    return np.array([[ux], 
                     [uy], 
                     [vx], 
                     [vy]])

def ppal(x):
    """Obtiene la polarizacion principal y equivalente a x"""
    
    u = x[:2,:]
    v = x[2:,:]
    x_p = np.zeros((4,1))

    if norm(u) != norm(v):
        
        aux = -2 * np.inner(u.flatten(), v.flatten()) / (norm(v)**2 - norm(u)**2)
        t_p = 0.5 * np.arctan(aux)
        u_p0 = u * np.cos(t_p) + v * np.sin(t_p)
        u_p1 = v * np.cos(t_p) - u * np.sin(t_p)
        u_p2 = -u_p0
        u_p3 = -u_p1

        v_p0 = u_p1
        v_p1 = u_p2
        v_p2 = u_p3
        v_p3 = u_p0 
        
        if norm(u_p0) > norm(v_p0) :
            if u_p0[1,0] > 0:
                x_p[:2,:] = u_p0
                x_p[2:,:] = v_p0
            elif u_p0[1,0] == 0:
                if u_p0[0,0] > 0:
                   x_p[:2,:] = u_p0
                   x_p[2:,:] = v_p0
                else:
                   x_p[:2,:] = u_p2
                   x_p[2:,:] = v_p2
            else:
                x_p[:2,:] = u_p2
                x_p[2:,:] = v_p2
        else:
            if u_p1[1,0] > 0:
                x_p[:2,:] = u_p1
                x_p[2:,:] = v_p1
            elif u_p1[1,0] == 0:
                if u_p1[0,0] > 0:
                   x_p[:2,:] = u_p1
                   x_p[2:,:] = v_p1
                else:
                   x_p[:2,:] = u_p3
                   x_p[2:,:] = v_p3
            else:
                x_p[:2,:] = u_p3
                x_p[2:,:] = v_p3
    
    else:
        if np.inner(u.flatten(), v.flatten()) == 0:
            u_p = norm(u) * np.array([[1],
                                      [0]])
            deter = u[0,0]*v[1,0] - u[1,0]*v[0,0]
            
            v_p = np.sign(deter) * norm(v) * np.array([[0],
                                                       [1]])
            x_p[:2,:] = u_p
            x_p[2:,:] = v_p
            
        else:
            t_p = np.pi / 4
            u_p0 = u * np.cos(t_p) + v * np.sin(t_p)
            v_p0 = v * np.cos(t_p) - u * np.sin(t_p)
            
            u_p1 = u * np.cos(t_p + np.pi/2) + v * np.sin(t_p + np.pi/2)
            v_p1 = v * np.cos(t_p + np.pi/2) - u * np.sin(t_p + np.pi/2)

            u_p2 = u * np.cos(t_p + np.pi) + v * np.sin(t_p + np.pi)
            v_p2 = v * np.cos(t_p + np.pi) - u * np.sin(t_p + np.pi)

            u_p3 = u * np.cos(t_p + 3*np.pi/2) + v * np.sin(t_p + 3*np.pi/2)
            v_p3 = v * np.cos(t_p + 3*np.pi/2) - u * np.sin(t_p + 3*np.pi/2)
            
            if norm(u_p0) > norm(v_p0) :
                if u_p0[1,0] > 0:
                    x_p[:2,:] = u_p0
                    x_p[2:,:] = v_p0
                elif u_p0[1,0] == 0:
                    if u_p0[0,0] > 0:
                       x_p[:2,:] = u_p0
                       x_p[2:,:] = v_p0
                    else:
                       x_p[:2,:] = u_p2
                       x_p[2:,:] = v_p2
                else:
                    x_p[:2,:] = u_p2
                    x_p[2:,:] = v_p2
            else:
                if u_p1[1,0] > 0:
                    x_p[:2,:] = u_p1
                    x_p[2:,:] = v_p1
                elif u_p1[1,0] == 0:
                    if u_p1[0,0] > 0:
                       x_p[:2,:] = u_p1
                       x_p[2:,:] = v_p1
                    else:
                       x_p[:2,:] = u_p3
                       x_p[2:,:] = v_p3
                else:
                    x_p[:2,:] = u_p3
                    x_p[2:,:] = v_p3

    return x_p


def T(theta):
    """Matriz de la transformacion en funcion de theta"""

    A = np.cos(theta)**2
    B = np.cos(theta) * np.sin(theta)

    T_theta = np.array([[1-A, -B, -A, -B], 
                        [-B, A, -B, A-1],
                        [A, B, 1-A, -B],
                        [B, 1-A, -B, A]])

    return T_theta


def Imagen(x):
    """Devuelve las coordenadas de los puntos de la imagen de la polarizacion x"""

    t = np.arange(0, 2*np.pi, dt)

    ux = x[0,0]
    uy = x[1,0]
    vx = x[2,0]
    vy = x[3,0]

    im_x = ux * np.cos(t) + vx * np.sin(t)
    im_y = uy * np.cos(t) + vy * np.sin(t)    

    return (im_x, im_y)


def show(r, alpha, beta, theta):

    alpha = np.radians(alpha)
    beta = np.radians(beta)
    theta = np.radians(theta)

    fig = plt.figure()
    fig.set_size_inches(15,15)
    
    scale = 1.5
    p0 = plt.subplot(121, aspect="equal", xlim=(-scale*r ,scale*r), ylim=(-scale*r, scale*r), title="Polarizacion In")
    p1 = plt.subplot(122, aspect="equal", xlim=(-scale*r, scale*r), ylim=(-scale*r, scale*r), title="Polarizacion Out")
    p0.grid()
    p1.grid()

    x = X(r, alpha, beta)
    p0.plot(*Imagen(x))
    ux, uy, vx, vy = ppal(x)
    p0.plot([0, ux[0]], [0, uy[0]])
    p0.plot([0, vx[0]], [0, vy[0]])
    
    y = np.matmul(T(theta), x)
    p1.plot(*Imagen(y))

    ux2, uy2, vx2, vy2 = ppal(y)
    p1.plot([0, ux2[0]], [0, uy2[0]])
    p1.plot([0, vx2[0]], [0, vy2[0]])
    
 
interact(show, r=fixed(1), alpha=(0., 180.0, 0.1) , beta=(-45., 45.0, 0.1), theta=(0., 180.0, 0.1))

```

```python

def r_a_b(x):

    x_p = ppal(x)
    u = x_p[:2,:]
    v = x_p[2:,:]
    r = norm(u)
    a = np.arctan2(u[1,0], u[0,0])
    deter = u[0,0]*v[1,0] - u[1,0]*v[0,0]
    b = np.sign(deter) * np.arctan2(norm(v), norm(u))

    return (r, np.degrees(a), np.degrees(b))

def show2(r, alpha, beta, theta, elev, azim):
    fig = plt.figure()
    fig.set_size_inches(15, 7)
    
    p0 = plt.subplot(121, xlim=(-10, 190), ylim=(-50, 50), title="Radio, Alpha, Beta In, Out y Trayectoria Out")
    p0.grid()
    p1 = plt.subplot(122, xlim=(-1, 5), ylim=(-10, 190), zlim=(-50, 50), projection='3d', title="Radio, Alpha, Beta In, Out y Trayectoria Out")


    alpha = np.radians(alpha)
    beta = np.radians(beta)
    theta = np.radians(theta)
    theta2 = np.arange(0, np.pi, dt)

    x = X(r, alpha, beta)
    y = np.matmul(T(theta), x)

    r_x, a_x, b_x = r_a_b(x)
    r_y, a_y, b_y = r_a_b(y)

    p0.scatter(a_x, b_x, color="green")
    p0.text(a_x, b_x, "  In: ({0:.2f}, {1:.2f}) ".format(a_x, b_x))

    p0.scatter(a_y, b_y, color="blue")
    p0.text(a_y, b_y, "  Out: ({0:.2f}, {1:.2f}) ".format(a_y, b_y))


    p1.scatter(*r_a_b(x), color="green")
    p1.text(*r_a_b(x), "  In: ({0:.2f}, {1:.2f}, {2:.2f}) ".format(*r_a_b(x)))

    p1.scatter(*r_a_b(y), color="blue")
    p1.text(*r_a_b(y), "  Out: ({0:.2f}, {1:.2f}, {2:.2f}) ".format(*r_a_b(y)))

    familia = np.array(list(map(lambda th : r_a_b(np.matmul(T(th), x)), theta2)))
    rs = familia[:,0]
    alphas = familia[:,1]
    betas = familia[:,2]

    p0.scatter(alphas, betas, color="orange")
   
    p1.scatter(rs, alphas, betas, color="orange")

    p1.view_init(elev, azim)



interact(show2, r=(0, 4, 0.1), alpha=(0.0, 180.0, 0.1) , beta=(-45.0, 45.0, 0.1), theta=(0., 180.0, 0.1), elev=(-90, 90), azim=(-180,180))
```

## Problema 1

Dadas dos polarizaciones $ \ X, Y \in \mathbb{P}, \ $ hallar, si es posible, un valor de $ \ \theta \ $ tal que $ \ T_\theta(X) \sim k \cdot Y, \ $ para algún valor de $ \ k \in \mathbb{R} $   

Sean $ \ (X)_B = \begin{pmatrix} u_x \\ u_y \\ v_x \\ v_y \end{pmatrix} \quad (Y)_B = \begin{pmatrix} p_x \\ p_y \\ q_x \\ q_y \end{pmatrix} \quad $ donde $ \ B, \ $ es la base canonica de $ \ \mathbb{P} $

Escribiendo $ \ a = \cos^2{\theta} \quad b = \sin{\theta} \cos{\theta} \ $ se tiene que :
$$ \lvert T_\theta \rvert_B = \begin{pmatrix} 1 - a &  -b   &  -a   &  -b   \\ 
                                               -b   &   a   &  -b   & a - 1 \\ 
                                                a   &   b   & 1 - a &  -b   \\ 
                                                b   & 1 - a &  -b   &   a      \end{pmatrix} $$
Teniendo en cuenta que :
$$ k \cdot Y \sim Y^\prime \iff \exists \ t_0 \ / \ Y^\prime = k \cdot \begin{pmatrix} p_x \cos{t_0} + q_x \sin{t_0} \\ 
                                                                                       p_y \cos{t_0} + q_y \sin{t_0} \\
                                                                                       q_x \cos{t_0} - p_x \sin{t_0} \\
                                                                                       q_y \cos{t_0} - p_y \sin{t_0} \end{pmatrix} $$
y escribiendo $ \ c = k \cos{t_0} \quad d = k \sin{t_0} \ $ podemos plantear lo siguiente :

$$ \begin{pmatrix} 1 - a & -b & -a & -b \\ -b & a & -b & a - 1 \\ a & b & 1 - a & -b \\ b & 1 - a & -b & a \end{pmatrix} \cdot \begin{pmatrix} u_x \\ u_y \\ v_x \\ v_y \end{pmatrix} = c \cdot \begin{pmatrix} p_x \\ p_y \\ q_x \\ q_y \end{pmatrix} + d \cdot \begin{pmatrix} q_x \\ q_y \\ -p_x \\ -p_y \end{pmatrix} $$

y reordenando:

$$ \begin{pmatrix} -u_x - v_x & -u_y - v_y & -p_x & -q_x \\
                    u_y + v_y & -u_x - v_x & -p_y & -q_y \\ 
                    u_x - v_x &  u_y - v_y & -q_x &  p_x \\
                    v_y - u_y &  u_x - v_x & -q_y &  p_y \end{pmatrix} \cdot \begin{pmatrix} a \\ b \\ c \\ d \end{pmatrix} = \begin{pmatrix} -u_x \\ v_y \\ -v_x \\ -u_y \end{pmatrix} $$

Luego el problema original tendrá solución, si la tiene este sistema lineal no homogeneo y además se cumple que $\ 0 \leq a \leq 1 \ $ y $ \ -\frac{1}{2} \leq b \leq \frac{1}{2} $
